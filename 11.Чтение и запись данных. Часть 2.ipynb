```python
session = req.urlopen(url)
response = session.read().decode()
```

```python
with open("weather.json", "w") as file:
    json.dump(json.loads(response), file, indent=4)
```

```python
with open("weather.json", "r") as file:
    data = json.load(file)

hourly_forecast = data["forecast"]["forecastday"][0]["hour"]
df = pd.DataFrame(hourly_forecast)
```
