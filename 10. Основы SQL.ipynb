### **Запросы**

Структура БД, запрос DESCRIBE

* describe - описание таблиц
* select - вывод
* where - что мы требует 
* order by - сортировка по возврастанию
* order by _ desc - сортировка по убыванию
* order by _ desc, _ asc - сортировка по воз-нию и убыванию
* limit (число) - количество, которое нам нужно вывести.
* select distinct - вывод маленького списка
* union (ALL) - вывод нескольких таблиц

### Выбор и фильтрация данных, запрос SELECT

> select name, type from courses; -> Вывод Имени и Типа из таблицы Курсов.

> select name from teachers where salary > 20000; -> Вывод Имени из таблицы Учителя, где Зарплата больше 20000.

> select * from students order by age desc; -> Вывод всех данных из таблицы студентов по убыванию по годам жизни.

> select * from teachers order by age desc, salary asc; -> Вывод всех данных из таблицы Учителя. По убыванию Года жизни, По возврастанию Зарплатую.
